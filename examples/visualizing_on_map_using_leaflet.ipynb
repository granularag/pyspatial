```python
import pandas as pd
from pyspatial import vector
from pyspatial.visualize import HTMLMap
```

```python
vl, vldf = vector.read_layer("../test/data/vector/cb_2014_us_state_20m.zip", index="STUSPS")

#Join some additional data at the state level
state_df = pd.read_csv("../test/data/state_info.csv")
state_df['State'] = state_df.State.str.title()
vldf = pd.merge(vldf, state_df, left_on='NAME', right_on='State')
vldf = vldf.set_index("STUSPS")

# Area of water
vldf["AWATER"] = vldf.AWATER/1.e6 # To square kmm

# Area of land
vldf["ALAND"] = vldf.ALAND/1.e6 # To square km

# Get a resonable center for the map
centroid = vl["KS"].Centroid()
lng, lat, z = centroid.GetPoint()
```

```python
# Specify what should go in the hover state for the choroplet
# Notice we can see data in for each shape, even though it 
# is not being visualized.  If not specified, it will use
# All the columns in the data frame
info_cols = ['NAME', 'AWATER', 'ALAND', 'Unemployment Rate', 'Governor',
             '2012 Presidential Election']

# Initialize the map
hmap = HTMLMap(lat, lng, data=vldf, info_cols=info_cols)

# Add the vector layer as the base layer of the map
hmap.set_baselayer(vl)

# Let's visualize some datda
hmap.choropleth("AWATER")
hmap.choropleth("Unemployment Rate")

# Can also choose palletes from here: http://colorbrewer2.org/
hmap.choropleth("ALAND", levels=7, palette="GnBu")

# Also handles categorical data with a custom pallete
# Here the I know the values in the column "Governor"
governors = {"Democratic": "#0000ff", "Republican": "#ff0000", "Independent": "#00ff00"};
hmap.choropleth("Governor", palette=governors)

# For fun, let's add the outline of CA, NY
hmap.add_shapes("shapes", vl[["CA", "NY"]])
hmap.render_ipython()
```
